```python
import math
```

```python
# ================================================== SIZE ================================================== #

def size():
  nominalDiameter = [15, 20, 25, 32, 40, 50, 65, 80, 100, 125, 150, 200, 250, 300, 350]

  while True:
    sizeInput = int(input('▷ size (unit; mm) \n\n   >>> '))
    try:
      if sizeInput in nominalDiameter:
        return sizeInput
    except ValueError:
      pass

result0 = size()

# ================================================ PRESSURE ================================================ #

def pressure():

  while True:
    pressureInput = input('\n▷ Max. Allowable Pressure (unit; MPa) \n\n   >>> ')
    if pressureInput.replace('.','',1).isdigit():
      pressureInput = float(pressureInput)
      return pressureInput
    break
    print('유효값을 입력해주세요.')

result1 = pressure()

# =============================================== MATERIAL ================================================= #

def material():

  material = {1:'Stainless Steel', 2:'Carbon Steel'}
  StainlessSteelMpa = {1:51, 2:47, 3:51, 4:47, 5:51, 6:48}  # 1:'304', 2:'304L', 3:'316', 4:'316L', 5:'A312 TP304', 6:'A276 316L'
  CarbonSteelMpa = {1:37, 2:110, 3:138, 4:161}   # 1:'SPPS250', 2:'A106 Gr.A', 3:'A106 Gr.B', 4:'A106 Gr.C'

  # ★★★ 2024.11.21 수정 : CS 2~4의 허용 인장 응력 수정, 상온 기준(ASME B31.3-2012 241page / 40 ℃) ★★★

  # StainlessSteel = {1:520, 2:480, 3:520, 4:480, 5:515, 6:485}  # 1:'304', 2:'304L', 3:'316', 4:'316L', 5:'A312 TP304', 6:'A276 316L'
  # CarbonSteel = {1:380, 2:330, 3:415, 4:485}   # 1:'SPPS250', 2:'A106 Gr.A', 3:'A106 Gr.B', 4:'A106 Gr.C'
  # 1 인장강도 unit >>> kgf/cm2
  # 2 인장강도 unit >>> Mpa(N/mm2)


  while True:
    try:
      materialA = int(input('\n▷ 사용 재질을 선택해주세요. (unit; Mpa) \n\n   (1) Stainless Steel\n   (2) Carbon Steel \n\n   >>> '))
      if materialA == 1:
        inputA = int(input(f'\n▷ {material[1]}의 소분류를 선택해주세요.\n\n   (1) 304 \n   (2) 304L \n   (3) 316 \n   (4) 316L \n   (5) A312 TP304 \n   (6) A276 316L \n\n   >>> '))
        if inputA in StainlessSteelMpa:
          return StainlessSteelMpa[inputA]
      elif materialA == 2:
        inputB = int(input(f'\n▷ {material[2]}의 소분류를 선택해주세요.\n\n   (1) SPPS250 \n   (2) A106 Gr.A \n   (3) A106 Gr.B \n   (4) A106 Gr.C \n\n   >>> '))
        if inputB in CarbonSteelMpa:
          return CarbonSteelMpa[inputB]
    except ValueError:
      pass

result2 = material()

# =============================================== CARCULATION ============================================== #

def calculation():
  # mas = result2 / 1.5   # 안전율 = 인장강도(극한강도, result2) / 재료 허용 응력(mas)
  result = 1000 * (result1 / result2)   # {최대 허용 압력(result1) / 재료 허용 응력(mas)} * 1000
  return result

schedule = calculation()

# =============================================== REFERENCES =============================================== #
'''
스케줄 수(번호) = {최고 사용압력(내부 작업 압력) / 재료의 허용 응력} * 1000
안전율 = 인장강도(극한 강도) / 재료의 허용 응력
'''
# ================================================ PRINT =================================================== #

print('\n\n==========================================================================')
print(f'결과 값 {schedule:.2f}(unit; mm) // {result0}A Pipe, SCH{round(schedule)} 이상을 사용하세요.')
print('==========================================================================')
```

* 일반적으로
150#, 300# -> SCH40 / 600#, 900# -> SCH80

* 특수한 경우
PMS에 Corrosion Arrowance가 보통은 1.5mm로 기재되어 있는데, 3mm인 경우가 있음.
이때는 SCH160 쓰기도 함.
